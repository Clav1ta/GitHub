## Übung 1 
Importiere requests, BeautifulSoup, Pandas  
Lest die Frontseite von SRF ein: http://ww.srf.ch  
Baut eine Liste mit den Titeln aller Artikel auf der Frontseite 
Speichert die Liste als .csv ab

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
requests.get('https://www.srf.ch/')
```

```python
r = requests.get('https://www.srf.ch/')
r.text
```

```python
# !!!!! WICHTIG !!!!!! r muss neu abgespeichert werden !!!!!

r = r.text
```

```python
soup = BeautifulSoup(r,"html.parser")
```

```python
# Titel auf der Frontseite verwenden den code <h3> </h3>
# Mit soup-Befehl alle h3-Titel hohlen

soup.find_all('h3')
```

```python
# wir haben die Liste mit allen Titeln erhalten. ABER es wird noch vieles mehr mitgeliefert.
# offenbar verwendet die Seite h3 nicht nur um Titel abzuspeichern
# 1.Schritt: Zahl der Elemente bestimmen

len(soup.find_all('h3'))
```

```python
# Testen was ist im ersten Element drin

soup.find_all('h3')[0]
```

```python
# Von Hand testen gehen, ab wann fangen die Titel an. Try and Error: [0] mit anderen Zahlen ersetzten zb [5:]

soup.find_all('h3')[10:]
```

```python
soup.find_all('h3')[13:]
```

```python
# Titel gibt es nur bis 55

lst = soup.find_all('h3')[13:][:55]
len(lst)
```

```python
for artikel in lst:
    # print(artikel)
    print(artikel.text.strip())
```

```python
# Warum die grosse Lücke im 2.Element?
lst[1].text.strip()
```

```python
for artikel in lst:
    art = " ".join(artikel.text.strip().split())
    print(art)
    
# Suchmaschine Duckduckgo "how to remove space python"
# ODER das fehlerhafte Element einfach weglöschen mit 
# liste = ['gutes element', '']
```

```python
fertigeliste = []
for artikel in lst:
    art = " ".join(artikel.text.strip().split())
    fertigeliste.append(art)
```

```python
fertigeliste
```

```python
pd.DataFrame(fertigeliste)
```

```python
pd.DataFrame(fertigeliste).to_csv('Alle_Titel_SRF-Frontseite.csv')
```

```python
# ODER VORSCHLAG VON LEA:
# sucht nach 'span' und class statt einfach nach h3

srftitel = []
for titel in soup.find_all('span', class_="title"):
    srftitel.append({"Titel":titel.text.strip()})
pd.DataFrame(srftitel).to_csv('SRF_Storytitel.csv')
```
