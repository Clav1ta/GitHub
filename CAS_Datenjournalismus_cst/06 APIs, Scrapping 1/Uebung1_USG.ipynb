# USGS-Abfrage 2

Abfrage von verschiedenen Zeiträumen mittels eines Loops

```python
lst = ["2018-07-01", "2018-08-01", "2018-09-01"]
lst2 = ["2018-07-31", "2018-08-31", "2018-09-30"]

for date, date2 in zip(lst, lst2):
    print(date,date2)
```

```python
basisurl = "https://earthquake.usgs.gov/fdsnws/event/1/"
was = "count?"
#"starttime=2018-09-01&endtime=2018-10-01": ursprüngliche Start- und Endzeit

for date, date2 in zip(lst, lst2):
    count = requests.get(basisurl+was+"starttime="+date+"&endtime="+date2)
    print(basisurl+was+"starttime"+date+"&endtime"+date2)
    print(count.json())  
    
```

# USGS-Abfrage 3

Wie lese ich die abgefragten Daten als pdf aus? 1.Schritt: gehe zu query-Abfrage: API  https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2014-01-01&endtime=2014-01-02

```python
basisurl = "https://earthquake.usgs.gov/fdsnws/event/1/"
was = "query?"
zeitrahmen = "starttime=2018-09-01&endtime=2018-10-01"
ft = "format=geojson&"    #format ist wie zB sum von Python belegt und kann nicht als Variable gebraucht werden
url = basisurl+was+ft+zeitrahmen
url

# url die es ausgibt ist ein String
```

```python
ergebnis = requests.get(url)
ergebnis.json()
```

```python
erdbeben = ergebnis.json()

# Abfrage was für ein Dateityp ist erdbeben: Ein Dictionary

type(erdbeben)
```

```python
# Abfrage was für Keys sind im Dictionary enthalten?

erdbeben.keys()
```

```python
#Ausgabe der Metadaten der Abfrage

erdbeben['metadata']
```

```python
# Abfrage der Metadaten zu den einzelnen Erdbeben

erdbeben['features']
```

```python
# Die einzelnen Variablen zu den einzelnen Ergebnissen können nachgesehen werden in der API Documentation.
# Uns interessiert nun die Magnitude und den Ort der einzelnen Erdbeben, einmal für die ersten 10.

# Das erste Element auslesen um einmal alle Features zu sehen. das wird als Dictonary ausgegeben
erdbeben['features'][0]
type(erdbeben['features'][0])
```

```python
erdbeben['features'][0]['properties']['mag']
```

```python
erdbeben['features'][0]['properties']['place']
```

```python
# For-Loop bauen wo man für alle 9000 Erdbeben die Magnitude ausliest
# MANIPULATION VON LISTEN
# lst [0] : nimmt das erste Element der Liste
# lst[:5] : die ersten "5" Elemente. : vor der Zahl bedeutet "bis zu".

for quake in erdbeben['features'][:10]:
    print(quake['properties']['place'], quake ['properties']['mag'])
```

```python
# Eigenes Dictionary bauen wo der Ort und Magnitude drin sind. Das dann abspeichern.
# Schritt 1: zwei Variablen definieren: Ort und Stärke.
# Schritt 2: Für jede Zeile ein neues, eigenes Dictionry bauen ( minidict).

for quake in erdbeben['features'][:10]:
    ort = quake['properties']['place']
    stärke = quake['properties']['mag']
    
    minidict = {'Ort': ort, 'Stärke': stärke}
    print(minidict)
    
```

```python
# aus den Dictionarys eine Liste erstellen über alle 9000 Elemente
# zuerst eine leere Liste definieren, die gefüllt wird.
# [:10] entfernen, da man

leerlist = []
for quake in erdbeben['features']:
    ort = quake['properties']['place']
    stärke = quake['properties']['mag']
    
    minidict = {'Ort': ort, 'Stärke': stärke}
    leerlist.append(minidict)
    
# ZUR KONTROLLE:
#leerlist ODER print(leerlist)
```

```python
# Variable umbenennen
volle = leerlist
volle
```
