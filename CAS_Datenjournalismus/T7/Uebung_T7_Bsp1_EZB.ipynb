# Übung EZB

Wir entwickeln einen kleinen Scraper der auf der Frontseite der EZB die drei Zinskurse täglich abgreift.  


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Funktion request verlangt IMMER einen String

r = requests.get("https://www.ecb.europa.eu/home/html/index.en.html")
print(r)
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
# Schreibweise siehe: https://www.crummy.com/software/BeautifulSoup/bs4/doc/ 

percentage = soup.find_all('td', {'class': 'stats-table-percentage'})
```

```python
percentage
```

```python
type(percentage)
```

```python
#Checken ob die Variable percentage den Wert richtig ausgibt

percentage[0].text.replace('\xa0','')
```

```python
percentage[1].text.replace('\xa0','')
```

```python
title = soup.find_all('td', {'class': 'stats-table-figure'})
```

```python
type(title)
```

```python
#Checken ob die Variable title den Titel richtig ausgibt

title[0].text
```

```python
percentagelist = []
# Hier beginnt der For Loop
for e, t in zip (percentage, title):
    # Hier lese ich aus beiden Listen die Elemente heraus
    # das e Element aus der Percentage-Liste berücksichtigt nur den Text
    # und ersetzt die Zeichenfolge xa0 mit nichts
    # Das t element zieht nur den Text aus dem HTML Tag.
    # Daraus wird eine Mini-Dictionary
    minidict = {'Title': t.text,
                'Percentage': e.text.replcace('\xa0','')}
    
    #
    percentagelist.append(minidict)
```

```python
# Diese Dictionaryliste wird in ein csv umgewandelt und abgespeichert.
pd.DataFrame(percentagelist).to_csv('percentage.csv')
```
